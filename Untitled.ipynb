```python
df = pd.concat(map(pd.read_csv, ['train.csv', 'test.csv']), ignore_index=True)
```

```python
df
```

```python
df = df.drop(['PassengerId', 'Name'], axis=1)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in range(11):
    if df.iloc[:,i].dtypes == 'O':
        x = df.iloc[:,i].values.reshape(-1,1)
        df.iloc[:,i] = le.fit_transform(x)
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
x = df['Age'].values.reshape(-1,1)
df['Age'] = imp.fit_transform(x)
df.isnull().sum()
```

```python
arr = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
for i in arr:
    x = df[i].values.reshape(-1,1)
    df[i] = imp.fit_transform(x)
df.isnull().sum()
```

```python
train = df.iloc[:8693,:]
test = df.iloc[8693:,:]
```

```python
test = test.drop(['Transported'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x = train['Transported'].values.reshape(-1,1)
train['Transported'] = le.fit_transform(x)
train.info()
```

```python
y = train['Transported']
x = train.drop(['Transported'], axis=1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()

x_train_min = mms.fit_transform(x_train)
x_test_min = mms.transform(x_test)
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()

x_train_norm = std.fit_transform(x_train)
x_test_norm = std.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score


knn = make_pipeline( PCA(n_components=8), KNeighborsClassifier(n_neighbors=18))
knn.fit(x_train_min, y_train)
pred_test_std = knn.predict(x_test_min)
print(accuracy_score(y_test, pred_test_std))
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train, y_train)
pred_test_std = knn.predict(x_test)
accuracy_score(y_test, pred_test_std)
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train_min, y_train)
pred_test_std = knn.predict(x_test_min)
accuracy_score(y_test, pred_test_std)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=47)
clf.fit(x_train_min, y_train)
pred = clf.predict(x_test_min)
print(accuracy_score(y_test, pred))
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train_norm, y_train.values.ravel())
Y_pred = lr.predict(x_test_norm)
print(accuracy_score(y_test, Y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x_train_norm, y_train)
Y_pred = dtc.predict(x_test_norm)
print(accuracy_score(y_test, Y_pred))
```
